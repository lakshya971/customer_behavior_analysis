```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))

# WE HAVE USED MEDIAN TO FILL NULL VALUES BECAUSE MEAN VALUES CAN EFFECT THE DATA QUALITY AS THEY ALSO CALCULATE OUTLIERS
# INSTEAD WE HAVE USED MEDIAN BECAUSE IT IS ROBUST w.r.t OUTLIERS AND DOES NOT AFFECT DATA QUALITY
```

```python
df.isnull().sum()
```

```python
# NOW WE KNOW THE COLUMNS NAMES INCLUDE BOTH CAPITAL AND LOWER CASES LETTERS SO IT WILL BE DIFFICULT FOR US TO REMEMBER THE EXACT FORMAT
# THAT'S WHY WE USE SNAKECASING (REMOVING BLANK SPACES WITH UNDERSCORES AND UPPERCASES INTO LOWERCASE)
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns = {"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

```python
# CREATE A COLUMNS AGE_GROUP

labels = ['Young_adult','Adult','Middle_age','Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)  

# QCUT IS USED TO DISTRIBUTE THE LABELS INTO FOUR EQUAL PARRTS
```

```python
df[['age','age_group']].head()
```

```python
# CREATE PURCHASE_FREQUENCY_DAYS

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every-3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
